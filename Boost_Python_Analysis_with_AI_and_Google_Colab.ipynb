```python
# prompt: I have a data frame called df.Can you show me what columns are in the dataframe?Also, for numeric columns , show me the distribution of the data. For categorical variables, show me the unique values and their frequencies.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is already defined from the previous code block
# Display column names
print("Columns in the DataFrame:")
print(df.columns)

# Analyze numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns
for col in numeric_cols:
  print(f"\nDistribution of {col}:")
  print(df[col].describe())  # Summary statistics
  plt.figure(figsize=(8, 6))
  sns.histplot(df[col], kde=True)
  plt.title(f'Distribution of {col}')
  plt.show()

# Analyze categorical columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    print(f"\nUnique values and frequencies for {col}:")
    print(df[col].value_counts())
```

```python
# prompt: 1. Can you analyze df for missing data? Please:
# * # 1. Show me the count and percentage of missing values for each column
# * # 2. Visualize the missing data patterns with a heatmap
# * # 3. Suggest appropriate strategies for handling the missing values based on their patterns
# * # 4. Identify any columns that might need to be dropped due to excessive missing data (>50%)

import pandas as pd
import matplotlib.pyplot as plt
# 1. Count and percentage of missing values
missing_counts = df.isnull().sum()
missing_percentages = (missing_counts / len(df)) * 100
missing_data = pd.DataFrame({'Count': missing_counts, 'Percentage': missing_percentages})
print("Missing Data Summary:")
print(missing_data)

# 2. Visualize missing data patterns with a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Data Heatmap')
plt.show()

# 3. Suggest strategies for handling missing values
print("\nMissing Value Handling Strategies:")
for col in missing_data.index:
    if missing_data.loc[col, 'Percentage'] > 0:
        print(f"- {col}: {missing_data.loc[col, 'Percentage']:.2f}% missing")
        if df[col].dtype in ['int64', 'float64']:
            print(f"  - Possible strategies: Imputation with mean/median, or more sophisticated methods like KNN imputation.")
        else:
             print(f"  - Possible strategies: Imputation with mode, or consider treating as a separate category ('Unknown').")

# 4. Identify columns to drop
columns_to_drop = missing_data[missing_data['Percentage'] > 50].index
print("\nColumns to consider dropping due to excessive missing data (>50%):")
columns_to_drop
```

```python
# prompt: I've identified outliers in my 'runtime' column using the IQR method. Rather than capping at percentiles, I'd like to completely remove rows where runtime is beyond 1.5*IQR from the quartiles.Show me the distribution of the dataset before and after removing these outliersCalculate and display key statistics (mean, median, std) for both original and cleaned datasetsCreate a box plot comparing both datasets side by sideWhat percentage of the data was removed as outliers?

import matplotlib.pyplot as plt
# Calculate Q1, Q3, and IQR for 'runtime'
Q1 = df['runtime'].quantile(0.25)
Q3 = df['runtime'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_cleaned = df[(df['runtime'] >= lower_bound) & (df['runtime'] <= upper_bound)]

# Calculate and display key statistics
print("Original Dataset Statistics:")
print(df['runtime'].describe())

print("\nCleaned Dataset Statistics:")
print(df_cleaned['runtime'].describe())

# Create box plots
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y=df['runtime'])
plt.title('Original Runtime Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(y=df_cleaned['runtime'])
plt.title('Cleaned Runtime Distribution')

plt.tight_layout()
plt.show()

# Calculate the percentage of data removed
outlier_percentage = ((len(df) - len(df_cleaned)) / len(df)) * 100
print(f"\nPercentage of data removed as outliers: {outlier_percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt # Correct the typo in the import statement from 'matplotlib.pyplott' to 'matplotlib.pyplot'

# Create 'length_category'
def categorize_length(runtime):
    if runtime < 40:
        return 'Short'
    elif 40 <= runtime <= 120:
        return 'Standard'
    else:
        return 'Long'

df['length_category'] = df['runtime'].apply(categorize_length)

# Create 'rating_tier'
def categorize_rating(rating):
    if rating <= 4.0:
        return 'Poor'
    elif 4.1 <= rating <= 6.5:
        return 'Average'
    elif 6.6 <= rating <= 8.0:
        return 'Good'
    else:
        return 'Excellent'

df['rating_tier'] = df['imdb_rating'].apply(categorize_rating)

# Show counts for each category
print(df['length_category'].value_counts())
print(df['rating_tier'].value_counts())

# Visualizations
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.countplot(x='length_category', data=df)
plt.title('Movie Length Categories')

plt.subplot(1, 2, 2)
sns.countplot(x='rating_tier', data=df)
plt.title('Movie Rating Tiers')

plt.tight_layout()
plt.show()

# Relationship between length category and rating tier
plt.figure(figsize=(8, 6))
sns.countplot(x='length_category', hue='rating_tier', data=df)
plt.title('Relationship between Movie Length and Rating')
plt.show()
```

```python
# prompt: Give me a line chart for average runtime by release_year. Return the chart only for releasE_year 1990 and afterAdd a smoothing line that returns a 5 year running average.

import matplotlib.pyplot as plt
# Filter data for release years 1990 and after
df_filtered = df[df['release_year'] >= 1990]

# Group by release year and calculate the mean runtime
runtime_by_year = df_filtered.groupby('release_year')['runtime'].mean()

# Calculate the 5-year rolling average
rolling_avg = runtime_by_year.rolling(window=5, center=True).mean()

# Create the line chart
plt.figure(figsize=(12, 6))
plt.plot(runtime_by_year.index, runtime_by_year.values, label='Average Runtime')
plt.plot(rolling_avg.index, rolling_avg.values, label='5-Year Rolling Average', linestyle='--')
plt.xlabel('Release Year')
plt.ylabel('Average Runtime')
plt.title('Average Movie Runtime by Release Year (1990 onwards)')
plt.legend()
plt.grid(True)
plt.show()
```
